```python
# Step 1: Import libraries and load the dataset

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Display settings
pd.set_option('display.max_columns', None)
pd.set_option('display.float_format', lambda x: '%.3f' % x)

# Load the dataset
train_path = "LargeTrain.csv"  # Adjust path as needed
test_path = "LargeTest.csv"

# Use low_memory=False to avoid dtype guessing issues
train_df = pd.read_csv(train_path, low_memory=False)
test_df = pd.read_csv(test_path, low_memory=False)

# Check the basic shape and preview the data
print("Train shape:", train_df.shape)
print("Test shape:", test_df.shape)

train_df.head()
```

```python
# Step 2: Dataset overview and basic statistics

# Check basic info
print("=== Train Data Info ===")
train_df.info()

# Check shape
print("\n=== Shape of Train Data ===")
print(train_df.shape)

# Describe numeric features
print("\n=== Statistical Summary ===")
print(train_df.describe())

# Correlation matrix
print("\n=== Correlation Matrix ===")
correlation_matrix = train_df.corr(numeric_only=True)

# Display heatmap
plt.figure(figsize=(16, 10))
sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Keep only numeric columns
numeric_df = train_df.select_dtypes(include=['int64', 'float64'])

# Drop constant columns (std=0)
numeric_df = numeric_df.loc[:, numeric_df.std() > 0]
```

```python
# Improved Correlation Heatmap

# Select numeric columns only
numeric_df = train_df.select_dtypes(include=['int64', 'float64'])

# Drop constant columns (zero standard deviation)
numeric_df = numeric_df.loc[:, numeric_df.std() > 0]

# Compute full correlation matrix
corr_matrix = numeric_df.corr().abs()

# Find upper triangle mask (to remove duplicates)
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

# Flatten, sort, and get top correlated pairs
corr_pairs = corr_matrix.where(mask).stack().sort_values(ascending=False)

# Get top 20 most strongly correlated feature pairs
top_corr_features = corr_pairs.head(20)

print("Top 20 Strongest Correlated Feature Pairs:\n")
print(top_corr_features)

# Create a heatmap for the features involved
top_features = list(set(top_corr_features.index.get_level_values(0).tolist() +
                        top_corr_features.index.get_level_values(1).tolist()))

top_corr_matrix = numeric_df[top_features].corr()

# Plot heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(top_corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Top Correlated Features (No Target Specified)")
plt.show()
```

```python
# Step 3: Missing Values Analysis

# Count of missing values per column
missing_values = train_df.isnull().sum()
missing_percent = (missing_values / len(train_df)) * 100

# Combine into a DataFrame
missing_df = pd.DataFrame({
    'Missing Values': missing_values,
    'Percentage (%)': missing_percent
})

# Filter only columns with missing data
missing_df = missing_df[missing_df['Missing Values'] > 0]
missing_df = missing_df.sort_values(by='Percentage (%)', ascending=False)

# Display
print("=== Missing Values Summary ===")
print(missing_df)

# Plot missing values
plt.figure(figsize=(12, 6))
sns.barplot(x=missing_df.index, y='Percentage (%)', data=missing_df)
plt.xticks(rotation=90)
plt.title('Missing Data by Column')
plt.tight_layout()
plt.show()
```

```python
# Check for missing values
total_missing = train_df.isnull().sum().sum()
print(f"🔍 Total Missing Values: {total_missing}")

# Check for duplicates
total_duplicates = train_df.duplicated().sum()
print(f"🔍 Total Duplicate Rows: {total_duplicates}")
```

```python
missing_per_column = train_df.isnull().sum()
missing_per_column = missing_per_column[missing_per_column > 0]
print("🔍 Columns with Missing Data:\n", missing_per_column)
```

```python
# Step 4: Unique Values & Cardinality

# Count unique values per column
unique_counts = train_df.nunique().sort_values(ascending=False)

# Create a DataFrame for easier viewing
cardinality_df = pd.DataFrame({
    'Column': unique_counts.index,
    'Unique Values': unique_counts.values
})

# Display top 20 high-cardinality columns
print("🔍 Top 20 Columns by Cardinality:\n")
print(cardinality_df.head(20))

# Optional: Plot unique value distribution
plt.figure(figsize=(14, 6))
sns.histplot(unique_counts, bins=50, kde=False)
plt.title("Distribution of Unique Values per Column")
plt.xlabel("Number of Unique Values")
plt.ylabel("Count of Columns")
plt.tight_layout()
plt.show()
```

```python
# Identify high-cardinality columns
high_cardinality = cardinality_df[cardinality_df['Unique Values'] > 5000]
print("🔎 High-Cardinality Columns (Possible IDs, Timestamps, etc.):")
print(high_cardinality)
```

```python
# Step 5A: Detect outliers using IQR method

numeric_cols = train_df.select_dtypes(include=['int64', 'float64']).columns
outlier_summary = {}

for col in numeric_cols:
    Q1 = train_df[col].quantile(0.25)
    Q3 = train_df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = ((train_df[col] < lower) | (train_df[col] > upper)).sum()
    outlier_summary[col] = outliers

# Convert to DataFrame
outlier_df = pd.DataFrame.from_dict(outlier_summary, orient='index', columns=['Outlier Count'])
outlier_df = outlier_df.sort_values(by='Outlier Count', ascending=False)

print("🔍 Top 20 Features with Most Outliers:\n")
print(outlier_df.head(20))
```

```python
# Step 5B: Boxplots for top outlier-heavy features

top_outlier_cols = outlier_df.head(6).index  # You can change number as needed

plt.figure(figsize=(16, 10))
for i, col in enumerate(top_outlier_cols, 1):
    plt.subplot(2, 3, i)
    sns.boxplot(y=train_df[col], color='skyblue')
    plt.title(f'Boxplot: {col}')
    plt.tight_layout()
plt.show()
```
